## 1.1.0 Random Variables


We will begin today with the concept of random variable, which is essential to probability and statistics.

<p style="color:blue; text-align:center"><i>A <b>random variable</b> is (for the purposes of this class) an observable whose outcome may differ between experiments.</i></p>

For example, if you're measuring the length of a cube in your experiment, the length you measure is a random variable. This is because your measured results conceivably might vary, depending on how you take the measurement and the precision of the ruler, or even thermal expansion and contraction of the cube.

If you're measuring the number of snakes in a population, the number of snakes you count is also a random variable.

<i>Technical note:</i> A random variable represents any possible values that the measurement could produce. The actual measured results are not random variables; they are data points. We often represent random variables by capital letters and their measured values as lowercase letters.

There are two types of random variables, continuous and discrete. Discrete random variables can only take on discrete values (snake example), whereas continuous random variables can take on continuous values (cube example).

<!--end-block-->

<!--start-block-->
 
## 1.1.1 Probability Density Functions
 
Discrete random variables are associated with a probability mass function that defines the exact probability that the random variable will take a certain discrete value. More precisely,

<p style="color:blue; text-align:center"><i>The <b>probability mass function (PMF)</b> for a discrete random variable $R$ is a function $p(r)$ that maps from possible values $r$ of $R$ to the probability that $R=r$.</i></p>

<i>Corollary:</i> the sum of $p(r)$ over all possible values of $R$ is 1. Thus, $p(r)$ is unitless

For continuous probability distributions, we have instead a probability density function.

<p style="color:blue; text-align:center"><i>The <b>probability density funtion (PDF)</b> for a continuous random variable $R$ is a function $P(r)$ defined on possible values $r$ of $R$, such that the probability that $R$ is between $a$ and $b$ is given by</i></p>

$$p(a < R < b) = \int_{a}^{b} P(r) dr.$$

Here, we adopt the convention that lowercase $p$ represent probabilities, and uppercase $P$ denote probability densities.

<i>Corollary:</i> the integral of $P(r)$ over all possible values of $r$ is 1. Thus, $P(r)$ has units of $1/r$.

<!--end-block-->

## 1.3 Gaussian distribution


The most common probability distribution is the Gaussian or normal distribution, which represents many natural phenomena, especially in physics. The normal distribution for a random variable $X$ with mean  $\mu$ and standard deviation $\sigma$ is given by

$$P(x) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left(-\frac{1}{2}\frac{(x-\mu)^2}{\sigma^2}\right)$$

## 1.4 Joint Probability Density Functions

We can define joint probability density functions, e.g. $P(x,y)$, for multivariate probability distributions in a similar way. 

<p style="color:blue; text-align:center"><i>The <b>joint PDF</b> for two continuous random variables $X$ and $Y$ is a function $P(x, y)$ defined on possible values $x$ of $X$ and $y$ of $Y$, such that the probability that $X$ and $Y$ are in some region $R$ is given by</i></p>

$$p((X, Y)\in R) = \iint_{R} P(x, y) dx dy$$

We define the single variable PDFs $P(x)$ and $P(y)$ by integrating (or marginalizing) the extra variable from the joint PDF, i.e. 

$P(x) = \int_{-\infty}^{\infty} P(x,y) dy$ and $P(y) = \int_{-\infty}^{\infty} P(x,y) dx$

(Here, we implicitly assumed that the possible values of $X$ and $Y$ are the whole real number line.)

Now we can introduce another very useful term in statistics:

<p style="color:blue; text-align:center"><i>Two random varibales $X$ and $Y$ are <b>independent</b> from one another if the joint PDF factors into the product of the two single-variable PDFs:</i>

$$P(x, y) = P(x) P(y).$$
    
<!--end-block-->

<!--start-block-->

Given $k$ independent Gaussian random variables, the above definition of independence guarantees that the joint PDF is


$$P(x_1, x_2, ... x_k) = \prod_{i=1}^k P(x_i) = \frac{1}{ (2\pi)^{k/2}\prod_i \sigma_i} \exp \left(-\frac{1}{2}\sum_i \frac{(x_i- \mu_i)^2}{\sigma_i^2}\right)$$

## 1.5 Poisson Distributions

Another very common distribution in physics is the Poisson distribution:

<p style="color:blue; text-align:center"><i>For a sequence of events that are independent of each other and independent of time, a random variable representing the number of events occuring within a given interval obeys the <b>Poisson distribution</b>.</i></p>

Examples include 
- the number of radioactive decays within a material over one second 
- the number of photons detected from a distant supernova in one day
- the number of meteorite impacts on Earth in one year

Note that Poisson-distributed random variables are discrete.

The PMF of a Poisson-distributed random variable with unitless parameter $\lambda$ is

$$p(k)=\frac{\lambda^k e^{-\lambda}}{k!}.$$

You can view $\lambda$ as the average rate of events times the time interval over which $k$ events are detected.

## 1.6 Simulating a Poisson Process


Let's simulate a **Poisson process**, which is just a process leading to a set of events that
- Are independent of time, meaning that they are not more or less frequent depending on the time of detection
- Are independent of each other, meaning the probability distribution of one event is unaffected by the presence of other events

Then a random variable that counts these events occuring within an interval is Poisson-distributed.

## Optional excersize: Exponential distributions

For Poisson processes, we've seen that
* The probability distribution of each event's time of occurence is <b>uniform</b>
* The probability distribution of the number of events occuring within an interval is <b>Poisson</b>

However, we have not discussed the probability distribution of the time between events. It turns out to be exponential, with PDF

$$P(\Delta t)= \lambda e^{-\lambda \Delta t}.$$

<!--end-block-->

Why is the probability distribution of the time between independent events exponential?

<span style="opacity:0.5">Hint 2: The probability distribution has not changed after you've waited some time $T$ (why?). That means that whatever the PDF $P(\Delta t)$ is, we must have $P(\Delta t) \propto P(\Delta t + T)$. The constant of proportionality is independent of $\Delta t$, and this formula is true for all $T$. Find a general class of distributions that satisfy this property. Then find the only distribution which has the $\lambda$ as the expected value for $\Delta t$.</span>

<!--
#solution
Suppose you detect an event and immediately start a stopwatch. Then some time $T$ passes with no events detected. The events are assumed to be independent of time, so given that time $T$ has passed, the probability distribution for how much time will pass until the next event has not changed. That means that $P(\Delta t) \propto P(\Delta t + T)$, with a factor of proportionality set by how likely it was for you to be able to wait for time $T$ wthout detecting an event in the first place.

The only distributions that satisfy this property are $P(\Delta t) = Ae^{-\lambda_0 \Delta t}$, where $A$ is a normalization constant and $\lambda_0$ is some positive real number. Requiring that $\int_0^\infty P(\Delta t) d\Delta t = 1$, we have $A=\lambda_0$.
 
We need the expectation value of $P(\Delta t)$ to be the average time between events, which is $1/\lambda$. Since $\int_0^\infty d\Delta t \Delta t \lambda_0 e^{-\lambda_0 \Delta t} = 1/\lambda_0$, we must have $\lambda_0 = \lambda$. Thus, the probability distribution for the time between events $\Delta t$ is 
$$P(\Delta t) = \lambda e^{-\lambda \Delta t}.$$
